Ejercicio 2: Dado el diccionario datos = {'a':10, 'b':25, 'c':7, 'd':18}

a) Crea una lista con las claves cuyo valor sea mayor a 12.

```python
datos = {'a':10, 'b':25, 'c':7, 'd':18}
claves_m = [k for k, v in datos.items() if v > 12]
print(claves_m)
```

b) Crea un nuevo diccionario con esas claves y sus valores al cuadrado.

```python
nuevo_dict = {k: datos[k]**2 for k in claves_m}
print(nuevo_dict)
```

Ejercicio 3: Define una función 'resumen(lista)' que retorne un diccionario con:

longitud, suma, promedio, mínimo y máximo de la lista.


```python
def resumen(lista):
    return {
        'longitud': len(lista),
        'suma': sum(lista),
        'promedio': sum(lista) / len(lista),
        'minimo': min(lista),
        'maximo': max(lista)
    }
print(resumen([10, 20, 30, 40]))
```

Ejercicio 4: Usa map() para aplicar la función potencia(x, 3) a la lista [1,2,3,4,5].

```python
lista_numeros = [1, 2, 3, 4, 5]

#potencia al cubo
def potencia(base, exponente=3):
    return base ** exponente
#map
resultado = list(map(potencia, lista_numeros))
print(resultado)
```

Ejercicio 1: Crea una matriz 5×4 de enteros aleatorios entre 1 y 100 (seed=7).

```python
import numpy as np

np.random.seed(7)
A = np.random.randint(1, 101, (5, 4))
print(A)
```

a) Calcula la media, máximo y mínimo de cada columna.

```python
print("Med:", A.mean(0))
print("Max:", A.max(0))
print("Min:", A.min(0))
```

b) Calcula la suma acumulada de la primera fila.

```python
print(np.cumsum(A[0]))
```

c) Normaliza la matriz para que cada columna tenga valores entre 0 y 1

Fórmula: (x - min_col) / (max_col - min_col)


```python
min_col = A.min(0)
max_col = A.max(0)

norm = (A - min_col) / (max_col - min_col)
print(norm)
```

Ejercicio 4: Crea dos matrices 3×4 y concaténalas vertical y horizontalmente.


```python
M1 = np.random.randint(1,10,(3,4))
M2 = np.random.randint(1,10,(3,4))
print(M1)
print(M2)
```

Vertical

```python
vert = np.concatenate([M1, M2], axis=0)
print(vert)
```

Horizontal

```python
hori = np.concatenate([M1, M2], axis=1)
print(hori)
```

```python
part = np.vsplit(vert, 3)

for p in part:
    print(p)
```

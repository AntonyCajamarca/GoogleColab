#Slicing_SeleccionSubconjuntos

```python
vista = A[0, :] # vista
copia = A[0, :].copy() #copia independiente

vista[0] = 999 # modifica A
copia[0] = 888 # NO se modifica A

print(A[0, :])
print(np.shares_memory(A, vista))
print(np.shares_memory(A, copia))
```

#Indexación Booleana

```python
import numpy as np
datos = np.array([[5, 12, 3, 18],
                  [7, 2, 15, 9],
                  [11, 6, 20, 1]])
```

Se crea mascara booleana

```python
mascara = datos > 10
print(mascara)
```

Aplicar máscara (filtro)

```python
mayores = datos[mascara] #devuelve los que cumplen con la mascara booleana
print(mayores)
```

Combinacion de condiciones

```python
#en este caso seran los valores de 5 al 15
rango = datos[(datos >= 5) & (datos <= 15)]
print(rango)
```

Modificar elementos con mascara

```python
#En este se remplaza los valores menores a 5 remplazando a 0
copia = datos.copy()
copia[copia < 5] = 0
print(copia)
```

Condicional vectoriado

```python
#codigo np.where(cond, valor_si_True, valor_si_false)
resultado = np.where(datos > 10, datos, 0) #deja todos los numeros mayores de 10 y reeamplaza con 0
print (resultado)
```

#Reshape el Cambio de Forma

```python
v = np.arange(12)
print(v)
```

Basico de reshape

```python
M2x6 = v.reshape(2, 6) # 2 filas, 6 columnas
M3x4 = v.reshape(3, 4) #no se pueden poner del mismo numero 4,4
M2x2x3= v.reshape(2, 2, 3) # tensor 3-D
```

Uso de -1 la dimensión automática

```python
auto = v.reshape(-1, 3) #  12/3 = 4 Se pone donde esta el -1 (4,3)
print(auto.shape)
auto2 = v.reshape(2, -1) # 12/2 = 6 aqui es ek mismo caso (2,6)
print(auto2.shape)
```

flatten vs ravel

```python
M = v.reshape(3, 4) # no poner el mismo (4,4)
flat = M.flatten() # este le hace una COPIA
rav = M.ravel() # lo hace VISTA
flat[0] = 999
print(M[0, 0]) #  flatten no modifica el original
rav[0] = 777
print(M[0, 0]) #  ravel SÍ modifica el original (es vista)
```

Agregar dimensiones

```python
v1d = np.array([1, 2, 3, 4]) # shape (4,)
col = v1d.reshape(-1, 1) #  vector columna
fil = v1d.reshape(1, -1) # vector fila
print(col.shape, fil.shape)
print(v1d)
```

#Broadcasting

Escalar mas el vector

```python
v = np.array([1, 2, 3, 4])
print(v + 10) #Hace escalarmente
```

Vector mas matriz

```python
Ma = np.ones((3, 4)) # shape (3, 4)
ve = np.array([10, 20, 30, 40]) #  convierte a (1,4) → (3,4)
print(Ma + ve)
```

Columna mas fila

```python
col = np.array([[1], [2], [3]]) # shape (3, 1) hay tres filas y una columna
fil = np.array([10, 20, 30, 40]) # shape (4,) → (1, 4) en este caso hay una fila y cuatro columnas
suma = col + fil
print(suma)
```

Normalización por columna

```python
datos = np.random.randint(1, 100, (5, 3))
maximos = datos.max(axis=0) # shape (3,) o (1,3) — máximo de cada columna
norm = datos / maximos # broadcasting (5,3) / (3,) → (5,3)
print(norm.round(2))
```

Verificar compatibilidad

```python
# Shapes incompatibles: (3,4) y (3,) → Error (sin reshape)
try:
  err = np.ones((3,4)) + np.ones(3,)
except ValueError as e:
  print('Error:', e)
# Shapes compatibles: (3,4) y (4,) → OK
try:
  err = np.ones((3,4)) + np.ones(4,)
except ValueError as e:
  print('Error:', e)
```

#Estadística y Reducciones

```python
M = np.array([[4, 7, 2, 8],
              [3, 11, 5, 1],
              [6, 9, 0, 4]])
print(M)
```

 Reducción global

```python
print(M.sum())
print(M.mean())
print(M.max())
print(M.min())
```

Reducción por eje

```python
print(M.sum(axis=0)) # suma por columna
print(M.sum(axis=1)) # suma por fila
print(M.mean(axis=0)) # media de cada columna
print(M.max(axis=1)) # máximo de cada fila
```

argmin / argmax — índice del extremo

```python
print(M.argmin(axis=1)) # índice del mínimo en cada fila
print(M.argmax(axis=0)) # índice del máximo en cada columna
```

std y var

```python
print(M.std()) # desviación estándar global
print(M.var(axis=0)) # varianza de cada columna
```

keepdims sirve para mantener dimensiones

```python
media = M.mean(axis=0, keepdims=True) # shape (1, 4) en vez de (4,)
print(media.shape)
centrado = M - media # broadcasting funciona directamente
```

#Concatenación y Apilamiento de Arrays

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])
B = np.array([[7, 8, 9],
              [10, 11, 12]])
```

Concatenación o concatenate

```python
vertical = np.concatenate([A, B], axis=0) # apilar filas
print(vertical.shape)
horizontal = np.concatenate([A, B], axis=1) # apilar columnas
print(horizontal.shape)
```

vstack y hstack

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
print(np.vstack([v1, v2])) # [[1 2 3], [4 5 6]] que es el shape (2,3)
print(np.hstack([v1, v2])) # [1 2 3 4 5 6] que es el shape (6,)
```

Para crear nuevo eje o stack

```python
capas = np.stack([A, B], axis=0) # nuevo eje al frente
print(capas.shape)
```
